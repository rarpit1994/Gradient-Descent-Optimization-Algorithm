```python
import numpy as np
import pandas as pd
import scipy.stats as sc
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('housing.csv')
```

```python
data = data.drop(labels= ['ocean_proximity','total_bedrooms'],axis=1)
```

```python
    
    i = 0
    alpha = 0.01
    epsilon = 0.00001
    neg_log_like_loss = []
    iterations = []
    
    #sigma_cap = np.random.normal(loc=100, size=(len(data_matrix[0, :])))
    #mu_cap = randColVals(data_matrix)
    #initial_vals = np.array([mu_cap, sigma_cap])
    initial_vals = np.random.normal(loc=43, size=(2,8))

    while True:
        
        del_mu_cap = - np.sum(( data_matrix - initial_vals[0, :]), axis=0)/(initial_vals[1, :]**2)
        del_sigma_cap = (((data_matrix.shape[0])/initial_vals[1]) - (np.sum(((data_matrix - initial_vals[0, :])**2), 
                                                                      axis=0)/initial_vals[1, :]**3))
        
        del_f_matrix = np.array([del_mu_cap, del_sigma_cap]) 
        
        final_vals = initial_vals - (alpha * del_f_matrix)
        final_vals[1] = abs(final_vals[1])
        
        func_old_vals = -np.sum(np.log(sc.norm.pdf(data_matrix, initial_vals[0], initial_vals[1])), axis=0)
        func_new_vals = -np.sum(np.log(sc.norm.pdf(data_matrix, final_vals[0], final_vals[1])), axis=0)
        
        func_old_vector = np.sqrt(np.sum(func_old_vals**2))
        func_new_vector = np.sqrt(np.sum(func_new_vals**2))
        
        if abs(func_old_vector - func_new_vector) < epsilon:
            break
        
        print("The initial and final value of Neg Log Likelihood in iteration {} is {} and {}\n".format(i, func_old_vector,func_new_vector))
        
        initial_vals = final_vals
        i += 1
    
        iterations.append(i)
        neg_log_like_loss.append(func_old_vector)
```
